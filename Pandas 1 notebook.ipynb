**<code style="background:yellow;color:black">Pandas is an open-source data manipulation and analysis library</code> for the Python programming language. ### <code style="background:yellow;color:black">Why do we use pandas?</code> 

1. **<u>Data Import and Export:</u>** Pandas makes it easy to import data from various file formats, such as CSV, Excel, SQL databases, JSON, and more. You can also export data to these formats.

2. **<u>Data Cleaning and Preprocessing:</u>** Pandas offers a wide range of functions for data cleaning, transformation, and preprocessing. This includes handling missing data, removing duplicates, converting data types, and more.

3. **<u>Data Exploration and Analysis:</u>** You can perform various data analysis tasks with Pandas, such as filtering, sorting, grouping, aggregating, and computing descriptive statistics. It allows you to gain insights into your data quickly.

4. **<u>Time Series Analysis:</u>** Pandas has built-in support for time series data, making it well-suited for working with temporal data and conducting time-based analysis.

5. **<u>Data Visualization:</u>** While Pandas itself is not a data visualization library, it can be easily integrated with libraries like Matplotlib and Seaborn for data visualization.

6. **<u>Integration with Other Libraries:</u>** Pandas can seamlessly integrate with other Python libraries commonly used in data science, such as NumPy for numerical operations and scikit-learn for machine learning.

7. **<u>Performance and Efficiency:</u>** Pandas is designed for high performance and memory efficiency, making it suitable for working with large datasets.

* **<code style="background:yellow;color:black">pd.DataFrame()</code> is a function used to create a dataset from scratch**. 
* **So, far we have loaded the pre-existing dataset. But we want to create a table from scratch.** 
* What if we want to create a dataframe from scratch, how do we do that? 
* Lets say we want to create the below dataset: There are 3 rows and 6 columns, and we want to create this dataset from scratch.

|  |country     |year |population |continent |life_exp |gdp_cap    |
|:-|:-----------|:----|:----------|:---------|:--------|:----------|
|0 |Afghanistan |1952 |8425333    |Asia      |28.801   |779.445314 |

**<u>There are two approaches to create a DataFrame from pd.DataFrame() function:**</u>

**<code style="background:yellow;color:black">1. Row-oriented approach:</code>**

* **Its called row-oriented approach because we are providing one row at a time.** 
* **Within this pd.DataFrame() function, we have to provide 2 parameters.** 
* The 1st parameter is a 2D list. As we can see in below example, its a nested list where every row is represented with one list. There are 3 rows as it was required, we have created that. 
* Now for the 2nd parameter, first we put a comma and in 2nd parameter we give column names as a list. Column names are given in comma seperated sequence. 
* So if we run sample_df, we can see the created dataset.

```python
df1 = pd.DataFrame(
              [
                  ['Afghanistan', 1952, 8425333, 'Asia', 28.801, 779.445314 ],
                  ['Afghanistan', 1957, 9240934, 'Asia', 30.332, 820.853030 ],
                  ['Afghanistan', 1962, 102267083, 'Asia', 31.997, 853.100710 ]
              ],

             columns = ['country','year','population','continent','life_exp','gdp_cap']
    )

df1
```

**<code style="background:yellow;color:black">2. Column-oriented approach:</code>**

* **Here we are using same pd.Dataframe() function to create our own dataset, but we are providing dictionary as a parameter**. * We have covered our dictionary session, we know dictionary is a collection of key value pairs. 
* **Here, column name is the dict key and list of column values or rows is dict value**. 
* There is only one parameter to be provided in column oriented approach.

```python
df2 = pd.DataFrame(
              {
                  'country' : ['Afghanistan', 'Afghanistan'],
                   'year' : [1952,1957],
                  'population' : [842533, 9240934],
                   'continent' : ['Asia', 'Asia'],
                  'life_exp' : [28.801, 30.332],
                   'gdp_cap' : [779.445314, 820.853030]
               }
    )

df2
```

### <code style="background:yellow;color:black">4. Renaming columns:</code>

**<code style="background:yellow;color:black">rename() method</code> is used to rename column names.** 

* **To rename a column name, we pass a dictionary in the rename() method, where dict key is the existing column name and dict value is the new name that we want to update.** 
* Here below, we implemented that code, but column names did not get updated as we can see in output, because by default axis is 0, and axis 0 refers to vertical axis. So when we are not giving any axis, it will take default axis as 0; it is looking for column names over the indices where no column names are present, basically its searching vertically the column names to update.

* So above syntax is creating a copy of original dataframe which is containing the modified column names. Thats why when below we do df.head(), the original dataframe is not changed.
* **One way by which we can collect or store the copy of modified data is to collect it in a new variable called new_df as done below.**
* But again we see original dataframe is not updated by checking df.head().

* By default, inplace is false. Below we are implementing inplace = True. This doesnt return anything because its modifying the original dataframe. No copies being returned, unlike in the above code we saw that copy was returned as an output. 
* Now, if we check using df.head(), we see original dataframe is changed.

```python
df.head()
```

```python
df.rename(
    {
        "population" : "Population",
         "country" : "Country"
    },

    axis = 1,
    inplace = True
)
```

# <code style="background:yellow;color:black">Dataframe indexing:</code>

**<code style="background:yellow;color:black">DataFrame indexing</code> in pandas refers to the process of selecting specific rows and/or columns from a DataFrame. Indexing allows us to retrieve, filter, and manipulate data in a DataFrame based on certain criteria.** 

### <code style="background:yellow;color:black">Implicit and Explicit indexes:</code>

**There are two types of indexes in pandas dataframe: <code style="background:yellow;color:black">Implicit and Explicit.</code> In pandas, the terms "implicit index" and "explicit index" refer to different ways of labeling and referencing data within a DataFrame or Series.**

**<u><code style="background:yellow;color:black">Implicit index:</code></u>** 

* **An implicit index is the default integer-based index that is automatically assigned to each row in a DataFrame or element in a Series when the data structure is created.**
* Implicit index is same as python index. Python indexing starts from 0 right, so does implicit index.
* **<code style="background:yellow;color:black">Implicit index always starts from 0.</code> So, in backend it is always index 0,1,2,...**
* The implicit index is created automatically unless a custom index is specified.

**<u><code style="background:yellow;color:black">Explicit index:</code></u>** 

* **An explicit index is a user-defined set of labels assigned to each row or element.**
* Whatever we see here on screen is called **<code style="background:yellow;color:black">explicit index.</code>** 
* Explicit index can be any immutable datatype, such as integers, strings, or datetime objects.
* **An explicit index is specified by the user when creating the DataFrame or Series or by modifying the index later using methods like set_index() method.**

### <code style="background:yellow;color:black">How to access these indices or in other words how to access various rows:</code>

* Pandas have provided us with different ways to access implicit and explicit indices or rows of a DataFrame. 
* **There is an indexing called iloc which will be used to access implicit indices and whenever we want to access explicit indices, we use an indexing called loc**. 
* Easiest way to remember is - iloc word starts with i and implicit word starts with i as well. 
* **<code style="background:yellow;color:black">Implicit index --> 0 --> iloc</code>**
* **<code style="background:yellow;color:black">Explicit index --> whatever is visible --> loc</code>**

Now lets understand the use cases of these two.

# <code style="background:yellow;color:black">set_index() method:</code>

* **In pandas, the <code style="background:yellow;color:black">set_index() method</code> is used to set a specific column or a combination of columns as the DataFrame's index. The index of a DataFrame is like its "row labels" and helps in identifying and accessing rows more efficiently.**
* Setting an index can be beneficial for various operations, such as merging/joining DataFrames, grouping data, and performing certain types of data analysis.
* **With this method, we are basically setting the index to some other value.**
* **What this below code cell <code style="background:yellow;color:black">df.set_index("Country")</code> does is like--** I am saying consider country column as my index and store in a copied dataframe called temp. In the output of this it is clearly seen that row indices are changed to country names. **So row indices can be anything.** 
* **How to access Afghanistan row indices? Like the code below.** Because the row indices are visible on screen means they are explicit, we use loc. It will give us all the rows where its afghanistan. Again these are explicit indices, because implicit are always 0,1,2,3,...

```python
temp = df.set_index("Country")

temp.head()
```

```python
temp.loc["Afghanistan"]
```

```python
# df remains unchanged because we have changed indexes of temp dataframe.

df.head()
```

# <code style="background:yellow;color:black">reset_index() method:</code>

* **The <code style="background:yellow;color:black">reset_index() method</code> in pandas is used to reset the index of a DataFrame. When we set an index using the set_index() method or similar operations, the existing index is replaced with the specified column(s), and the DataFrame may no longer have a default integer-based index. The reset_index() method allows us to revert to the default integer-based index and move the current index back to regular columns.**
* With this methos we can reset the index back to original.**
* **In the below code cell, reset_index() method is used.** Notice that in the output of the code below, 0,1,2,3...1703 is resetted. And, original index 3,4,5,6,...1706 is created as a new column.
* **If we dont want that new index column of 3,4,5,6,...1706, we can implement the code below. What it does is it automatically resets and removes that index column as well. So it resets and removes unnecessary column.**
* Now, in code below to below cell "df.reset_index(drop = True)", it showed us the required output, but the code df.head() doesnt show us that required output. 
* Why? - Because we havent made it permanent by saying inplace = True. So now if we check df.head(), we get back the original dataset.

```python
df.reset_index()
```

```python
df.reset_index(drop = True)
```

```python
df.head()
```

```python
df.reset_index(drop = True, inplace = True)

df.head()
```

### <code style="background:yellow;color:black">Difference between pandas method and function:</code> 

* **<code style="background:yellow;color:black">When we have created a dataframe object "df" (or any other variable name), any function which is applied on top of this pandas dataframe object is a method.</code>** 
* df refers to a specific dataframe object on which we are applying a method. 
* These are typically methods that are called on an existing dataFrame object for example df.mean(), df.drop(), df.groupby(). 
* These methods operate directly on the dataFrame, and they can access and modify its data and structure.

<br>

* **<code style="background:yellow;color:black">When we say pd.any_function_name(), it refers to a function that has been written inside pandas module.</code>** 
* pd refers to the pandas library itself. 
* These are functions provided by the pandas library, which we call with dataFrame or series objects as arguments for example pd.read_csv(), pd.concat(), pd.merge(). 
* These functions usually take dataframe or series objects as input and return new objects or perform operations that don't modify the original dataFrame directly.
