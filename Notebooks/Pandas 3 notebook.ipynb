### <code style="background:yellow;color:black">loc accessor:</code>

* **In pandas, the <code style="background:yellow;color:black">.loc accessor</code> is used for label-based indexing. It is primarily used to select rows and columns from a DataFrame by their labels or a boolean array.**

* **The basic syntax for using .loc is: <code style="background:yellow;color:black">DataFrame.loc[row_indexer, column_indexer]</code>**

**<u>Here, row_indexer and column_indexer can take various forms:**</u>

* **<code style="background:yellow;color:black">Label-based selection:</code>** If we provide a single label for row_indexer, it selects the corresponding row. If we provide a single label for column_indexer, it selects the corresponding column. If we provide labels for both row and column indexers, it selects the intersection of the specified row and column. **Example:** Selecting a single row and multiple columns - **df.loc['row_label', ['column1', 'column2']]**

* **<code style="background:yellow;color:black">Slicing with labels:</code>** We can use label-based slicing to select a range of rows or columns. **Example:** Selecting a range of rows and columns - **df.loc['row_label1':'row_label2', 'column_label1':'column_label2']**

* **<code style="background:yellow;color:black">Boolean indexing:</code>** We can use boolean conditions to filter rows based on a certain criterion. **Example:** Selecting rows based on a condition - **df.loc[df['column'] > 5]**

* **<code style="background:yellow;color:black">Selecting all rows for specific columns:</code>** We can use "**:**" as a wildcard to select all rows for specific columns. **Example:** Selecting all rows for specific columns - **df.loc[:, ['column1', 'column2']]**

We have to keep in mind that .loc is inclusive on both ends when using label-based slicing, unlike integer-based slicing, which is exclusive on the upper bound.

<u>Here's a simple example:</u>

```python
# Creating a sample DataFrame
data1 = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}

df1 = pd.DataFrame(data1, index = ['X', 'Y', 'Z'])

# Using .loc to select specific rows and columns
selected_data = df1.loc[['X', 'Y'], ['A', 'B']]
selected_data
```

# <code style="background:yellow;color:black">apply() method:</code>

* **In pandas, the <code style="background:yellow;color:black">apply() method</code> is used to apply a function along the axis of a DataFrame or Series. The apply method is particularly useful when we want to perform a custom operation on the elements of a DataFrame or Series.**
* **To understand .apply(), lets do this task:** 
    * Task: Give encoding to the Genders: 
        * Male --> 0 
        * Female --> 1 

    * Lets write a function for this, which we have implemented in the below code cell. <br><br>

* **<code style="background:yellow;color:black">Now, we want to apply this function on the entire column.</code> So, we apply here .apply() method here in next code cells.** 
* This is similar to vectorization. <br><br>

* **<u><code style="background:yellow;color:black">Here's a step-by-step breakdown:</code></u>**

    1. The encode_gender function is defined, which takes a parameter "g" (presumably representing gender) and returns 0 if "g" is equal to "Male" and 1 otherwise.

    2. The apply method is used on the "gender" column of the DataFrame "data". For each value in the "gender" column, the encode_gender function is applied, resulting in the transformation of "Male" to 0 and any other value to 1.

    3. The transformed values are then assigned back to the "gender" column in the DataFrame data. <br><br>

* **So, after executing this code, the "gender" column in the DataFrame data will have values replaced with 0 for "Male" and 1 for any other gender.**

* **It will apply row by row that function and it will store the result back into gender column. We can create another column as well but, here we are modifying the same column.**

* In gender column, we have got encoding. All males have been given 0 and females 1.

* **<code style="background:yellow;color:black">This is how we can apply a particular function on a particular column.</code>**

**This kind of encoding is useful when we want to convert categorical variables into a format that can be used in machine learning models, especially models that require numerical inputs.**

```python
def encode_gender(g):
    
    if g == "Male":
        return 0
    
    else:
        return 1
```

```python
encode_gender("Male")
```

```python
data["gender"] = data["gender"].apply(encode_gender)

data.head()
```

# <code style="background:yellow;color:black">What is Grouping ? Group by in context of pandas</code>

**It could simply be understood through the terms -- split, apply, combine**

<img src="https://learning.oreilly.com/api/v2/epubs/urn:orm:book:9781491912126/files/assets/pyds_03in01.png" height = 350/>

1. **<code style="background:yellow;color:black">Split:</code> Breaking up and grouping** a DataFrame depending on the value of the specified key.

2. **<code style="background:yellow;color:black">Apply:</code>** Computing **some function**, usually an **aggregate, transformation, or filtering**, within the individual groups.

3. **<code style="background:yellow;color:black">Combine:</code> Merge the results** of these operations into an output array.

> Note: All these steps are to understand the topic

* Lets say we have a dataframe like the one shown in image above. Grouping can be understood in terms of split, apply and combine. 

* Lets say we want to group on the basis of "key" column. So what pandas does internally is it will first split the data "key-wise", so all the rows of A together, all the rows of B together and all the rows of C together. 

* Now, it will apply some aggregation like sum function. So, we are summing it up, so for all the A values we have sum as 5, for B 7 and for C 9. 

* And then finally, it will merge or combine. That means merge the results of the aggregate functions into an output array. So, if we see input and output just by looking, we can see its groupby and aggregation, we group by key and we get sum of each group. Same as SQL. 

* The beauty here is these things are running in parallel and this makes it very very fast. 

* How to implement in code, we will see below.

# <code style="background:yellow;color:black">groupby() method:</code>

* **The <code style="background:yellow;color:black">groupby() method</code> in pandas is used to split a DataFrame into groups based on some criteria, apply a function to each group independently, and then combine the results back into a DataFrame. It is a powerful and flexible tool for data analysis and manipulation.**
* The groupby() method is a key tool for analyzing and summarizing data based on different criteria, and it is commonly used in combination with other pandas functions to perform complex data manipulations and analyses.

```python
data.head()
```

**We want to apply groupby over director column. Simply the column on the basis of which we want to groupby. When we see the output of this, we see that it is a groupby object. Its not a dataframe. <code style="background:yellow;color:black">This groupby object provides us with multiple functionalities.</code>** 

```python
data.groupby("director_name")
```

* **The <code style="background:yellow;color:black">.ngroups attribute</code> in pandas is used to get the number of groups formed after applying the groupby() method on a DataFrame.** 
* When we use groupby() to group a DataFrame based on one or more columns, it creates distinct groups based on the unique values in those columns. 
* The .ngroups attribute returns the number of such groups. It tells us how many groups are there. 
* So in our example, it means there are 199 director names or groups. 
* Here, every director name will be one group.

```python
data.groupby("director_name").ngroups
```

* **We are lets say more interested to know how the groups are stored. We dont have to worry about how pandas does it internally.**
* **The <code style="background:yellow;color:black">.groups attribute</code> is often used to inspect the composition of groups after applying the groupby() method.**
* It shows us all the groups. 
* The output of this attribute is a dictionary. 
* First example says dict key is director name and dict values is a list of all the indices where the director name is Adam McKay.

```python
data.groupby("director_name").groups
```

* **The <code style="background:yellow;color:black">get_group() method</code> in pandas is used to retrieve a specific group from a grouped DataFrame. After using the groupby() method to create groups based on certain criteria, we can use get_group() to extract a particular group by specifying the key associated with that group.**
* It actually gives us that particular group's all rows. 
* Also note that all the indices were there in the output of .groups as well for Adam McKay.

# <code style="background:yellow;color:black">Multiple aggregations:</code>

**In pandas, when we perform groupby operations, we can apply <code style="background:yellow;color:black">multiple aggregation functions</code> to different columns using the agg() or aggregate() method. This allows us to compute various summary statistics for each group in a single step.**

### <code style="background:yellow;color:black">aggregate() or agg() method:</code>
    
* **In pandas, <code style="background:yellow;color:black">agg() and aggregate() methods</code> are essentially the same method with different names. They both serve the same purpose and can be used interchangeably. Both methods are used for aggregation operations in the context of a grouped DataFrame.**
* So, we can choose to use either agg() or aggregate() based on our preference.
* **The aggregate() method in pandas is used to apply one or more aggregation functions to the grouped data in a DataFrame. When we use groupby() to group a DataFrame based on one or more columns, we can use aggregate() to compute various summary statistics for each group.**

**<code style="background:yellow;color:black">Question:</code> For every director, find the first year and the last year in which they have directed a movie.**

* **Here we need to perform groupby on director's name and on year column we have to do two aggregations called min and max.** 
* When we are doing single aggregation we can simply write the aggregate method's name using the dot operator. 
* **But when we want to do multiple aggregations, we use this aggregate() or agg() method and then aggregate methods are passed in it as string parameters.**

```python
data.groupby("director_name")["year"].aggregate(["min", "max"])
```

**<code style="background:yellow;color:black">Quiz:</code> For each year , how you will get the average, lowest, and highest value of ratings of movies?**

* **This is another way or syntax to write aggregation wherein we provide a dictionary.**
* This syntax means that over the vote_average column apply all the 3 aggregations.
* **This is an advanced syntax where we can perform <code style="background:yellow;color:black">multiple aggregates on multiple columns simultaneously</code> by providing another key:value pair in the .agg() method.**

```python
data.groupby("year").agg(
    {
        "vote_average" : ["mean", "min", "max"]
    }
)
```

# <code style="background:yellow;color:black">Groupby filtering:</code>

* **<code style="background:yellow;color:black">Groupby filtering</code> in pandas involves applying a condition to groups of data and filtering out groups that do not satisfy the condition.**
* **We can use the groupby() method along with filtering method like filter() to achieve this.**
***
* Is the above question same as filtering those rows where budget is greater than 100 million? So if we see this question -- Display only those rows where budget > 100 M. Is this question same as the question we have solved just now? -- NO.
* **This question simply asks show me those rows where budget is more than 100 million. Whereas the above question asks if a director has done any one movie greater than 100 million, show me all their movies.** If this particular director James Cameroon for example, has done even one movie greater than 100 million, we will show all of his movies irrespective of their budgets.
* **This is called <code style="background:yellow;color:black">Groupby filtering.</code>**
* **Do we agree that either we are selecting all the rows of that director or none of the rows of that director? -- YES. Either select the entire group either reject the entire group.**
* **This concept is very similar to <code style="background:yellow;color:black">"groupby and having in sql".</code> Having will either select the entire group based on condition or will reject the entire group**
***
* **We have achieved this groupby filtering through various steps as above. But we can use one line method called filter() on top of groupby() to achieve groupby filtering with just one line of code. Earlier we were applying a function, this time we will filter the groups.**

### <code style="background:yellow;color:black">filter() method:</code>

* **In pandas, the <code style="background:yellow;color:black">filter() method</code> is used to subset a DataFrame based on a particular condition. The filter() method can be applied to both rows and columns, and its behavior depends on the type of criteria specified.**
* When used with groupby objects, the filter() method is particularly useful. It allows us to filter groups of data based on group-specific conditions.
***
* **Now in the code cell below, firstly data.groupby("director_name") will groupby director name.**
* **Now over that group, we are applying a filtering criteria.** 
* **What happens is that each group is treated like one dataframe, we are picking that dataframe and applying max() over it on the budget column, then we are checking is it greater than 100 or not.**
* **If this condition is true, we will filter in that group, if this condition is false we will filter out that group.**
* **Meaning is- either we are selecting the entire group based on condition or either we are rejecting the entire group. So, if max budget of that group is greater than 100, select that group, else reject that group.**
* Result is same as achieved above through various steps as discussed.
***
Very very important point is that **<code style="background:yellow;color:black">in loc we are filtering rows, and in filter() we are filtering groups.</code>**

```python
data.groupby("director_name").filter(lambda df: df["budget"].max() > 100)
```

**<code style="background:yellow;color:black">Question: Categorize a movie as risky or non risky.</code> A movie is risky if its budget is greater than the average revenue of all the movies of that director.**

* **Lets say a director has done 10 movies. For every movie, we have to tell whether that movie is risky or not.** 
* **<code style="background:yellow;color:black">So, we have to take average revenue of these 10 movies and we have to compare with budget of one movie.</code>**
* **And if the budget this one movie is greater than the average revenue of this director then its risky, if no then its not risky.**

**Which concept in sql is this similar to? -- <code style="background:yellow;color:black">WINDOW FUNCTION.</code> For every group, we want aggregated value of that row. We take the average of the entire window and then we use to compare aggregate for each row.**

* In below code cell, basically what has happened is, df["revenue"].mean() this aggregation is being performed on the revenue column of that group because first we are doing .groupby(). 
* So specifically, within that group, this average is being computed for each row.

```python
def check_risky(df):
    risky_flag = df["budget"] > df["revenue"].mean()
    df["risky"] = risky_flag
    return df
```

```python
result = data.groupby("director_name", group_keys = True).apply(check_risky)

result
```

**<code style="background:yellow;color:black">Approach 2: More simpler way</code>**

1. for every director - find the avg revenue
2. merge this with the original dataframe
3. row wise comparison

This is implemented below:

```python
# Calculate average revenue for each director
avg_revenue_by_director = data.groupby('director_name')['revenue'].mean().reset_index()
avg_revenue_by_director.rename(columns = {'revenue': 'avg_revenue'}, inplace = True)

# Merge with the original dataframe
data = pd.merge(data, avg_revenue_by_director, on = 'director_name', how = 'left')

# Create a 'risky' column based on the comparison
data['risky'] = data['budget'] > data['avg_revenue']

# Drop the 'avg_revenue' column if you don't need it in the final result
data.drop('avg_revenue', axis = 1, inplace = True)

data
```
