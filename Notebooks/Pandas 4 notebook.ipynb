# <code style="background:yellow;color:black">Multi-index DataFrame:</code>


* What if Steven Speilberg has been active and producing movies for most amount of his time? 
* Maybe the operation that we should perform is that -- **<code style="background:yellow;color:black">How many movies does a director direct per year.</code> And in this way we get to a more correct understanding of productivity.**
* Lets do that analysis, but while doing that analysis, we will learn something new called **<code style="background:yellow;color:black">"a table which has multiple indexes, or a table which is multi-indexed".</code>**
* So we want to groupby director name but we also want minimum year(the year they started), maximum year(the year they ended), title_count. 
* Here in this data structure, one column is subdivided into further columns. 
* **REFER NOTES FOR MORE CLARITY.** 
* In our header , we have 3 things- director_name, title_count and year. But year is further subdivided into min and max. So there are two levels of indexing - one is at the topmost layer which contains the year and within year, we contain another indexing for min and max.
***

* **For this we use the aggregate() function.** We will try to aggregate data. One way of aggregation is to see the minimum and the maximum values, we will use that. 

* Here in the code below we are **<code style="background:yellow;color:black">grouping by on director_name and aggregating on both year and title. Now on year and title, we want to run the aggregate() function, and in this function we will pass the dictionary.</code>**

* It reads as -- against year we want min and max and against title we want count. 
***

* **In pandas, <code style="background:yellow;color:black">a multi-index DataFrame</code> is a DataFrame with multiple levels of indices for both rows and columns. The DataFrame that we have created below is called a MULTI-INDEX DATAFRAME, because along the columns there are 2 sets of indexes or 2 levels of indexes.**
* This allows us to represent higher-dimensional data in a tabular format. 
* Multi-indexing is also known as hierarchical indexing. 
* It enables us to work with and analyze datasets with more complex structures than a standard two-dimensional DataFrame.
***

* Here in the example below we have first level of indexes as director_name, year and title. And within year we have min and max and within title we have count as secondary level of indexes.
* **Similar level of multi indexes can happen along rows and that will also be called a multi-index dataframe.** 
* The dataframe is still 2 dimensional, its just that the headers have subdivisions.
* In the example below, the director_name is used as the index column.
***

* One thing about pandas is that sometimes the commands can become too long. 

* So if we are gettiing confused, its always recommended to run those commands individually and then put those together, similar to what we did above- data.groupby(["director_name"])["title"].count() and then data.groupby(["director_name"])["title"].count().sort_values(ascending = False)

```python
data_agg = data.groupby(["director_name"])[["year", "title"]].aggregate(
    {
    "year": ["min", "max"],
    "title": "count"
    }
)

data_agg
```

* Now what if we dont want these multi level indexes? What if we want to restructure the existing data into a more simpler format? 

* We want single level index, such that **director_name will be as it is, year will not be subdivided into min and max, in fact, we will make only 2 primary level indexes out of it called "year_max" and "year_min". Similarly with title and count as "title_count".**
***

* We can convert the existing data_agg into a single-index dataframe. But that is very very cumbersome. So, what we can do is we can do the exact same aggregation process by which we created this "data_agg" dataframe, but a little differently. 

* **We will use aggregate() function, but this time <code style="background:yellow;color:black">instead of passing a dictionary, we are passing arguments.</code>**

* **First agrument will be "year_max" i.e. the column name desired and that we will assign a tuple of year (column that we want) and max (aggregator name) to it. Same we will do for remaining two.**

* **<code style="background:yellow;color:black">So when we use aggregate() function using the dictionary method, it generates a multi-indexed dataframe, but when we use it using passing arguments method, it generates a single-indexed dataframe.</code>**

* Now we have eliminated a fuss of a multi-indexed dataframe. 

* **There is no preference over single or multi-index dfs. Its upto us, our code and our company requirements.** 

```python
data_final = data.groupby("director_name")[["year", "title"]].aggregate(
    year_max = ("year", "max"), # Format -- new_col = ("col", "aggregator")
    year_min = ("year", "min"),
    title_count = ("title", "count")
)

data_final
```

**We can also do .reset_index(inplace = True) so that we get a numerical key index and a simple direct dataframe in which we have 4 columns and an index column seperately.**

```python
data_final.reset_index(inplace = True)

data_final
```

* **Lets suppose, we want 2 extra columns called "years_active" and "movies_active".** 

* Below is the code how we do that. 

* **We generated "year_active" and "movies_per_year" column based on existing columns.** 

```python
data_final["years_active"] = data_final["year_max"] - data_final["year_min"]

data_final
```

```python
data_final["movies_per_year"] = data_final["title_count"] / data_final["years_active"]

data_final
```

* **Now we are sorting using sort_values() method on "movies_per_year" column.**

* Now we can see Steven Speilberg is not at the top anymore, we now have a different director called Tyler Perry, who is directing more movies every single year. 
* This director was at 190th index. As we can see he has directed only 7 movies but at a much better or fast pace. Pace is higher, number of movies is lesser. 

```python
data_final.sort_values("movies_per_year", ascending = False)
```

# <code style="background:yellow;color:black">Pivot dataset:</code>

* **In the context of data analysis and reshaping data, the term <code style="background:yellow;color:black">"pivot"</code> is used to describe the transformation of data from one structure to another, often involving a rotation or rearrangement of the data.**

* **When we "pivot" data, we are typically changing the way it is organized or structured. In the context of spreadsheets or databases, pivoting might involve reorganizing data from rows to columns or vice versa. It's a way of looking at data from a different perspective.**

* The specific details of how data is pivoted can vary depending on the tools and methods used in a particular data analysis context.

* Now in our example dataframe above, lets suppose, we dont like the structure of the data. There are multiple time columns, and temperature and pressure of each date is in a seperate row. 

* What we want is restructure the data in a way in which every single time column becomes a different row. So we will change this data from a horizontal structure to a more vertical structure. How do we do that? 

* **There is something called melt() function in pandas.** Similar to excel pivot and tableau pivot.

### <code style="background:yellow;color:black">melt() function:</code>

* **In pandas, the <code style="background:yellow;color:black">melt() function</code> is used to transform or reshape a DataFrame.**
* **<code style="background:yellow;color:black">It essentially converts a DataFrame from a wide format to a long format,</code>making it more suitable for analysis or visualization.**

* **What melting means is it will melt the column into the rows.** 

* **The <code style="background:yellow;color:black">basic syntax</code> of the melt function is: pd.melt(frame, id_vars = None, value_vars = None, var_name = None, value_name = 'value', col_level = None)**

    * frame: The DataFrame we want to melt.
    * id_vars: Columns to be retained as identifier variables (not melted).
    * value_vars: Columns to be melted. If not specified, all columns not set in id_vars will be melted.
    * var_name: Name to be used for the 'variable' column (default is 'variable').
    * value_name: Name to be used for the 'value' column (default is 'value').
    * col_level: If columns are multi-level, use this to specify the level to melt (default is None).

* Below is the melt operation done and this is how our dataset looks like. **Our 3 untouched columns are still there. But now it has created 2 new columns. This variable column is the column labels of time that have been melted and this value column is the corresponding values within those original time columns.**

```python
pd.melt(data, id_vars = ["Date", "Drug_Name", "Parameter"])
```

* One thing which we can do is -- We can give column "variable" and "value" which is a standard that pandas provide everytime, to our own custom name.

* We have to perform the same operation of melt again. Or we could have done this initially means in the above code by providing some extra parameters. But we are doing it below. Basically in the same melt operation only we can change the column variable and value names according to our requirement.

* Check here that we pass two parameters: <code style="background:yellow;color:black">var_name = "Time" and value_name = "Reading".</code>

```python
data_melt = pd.melt(
    data,
    id_vars = ["Date", "Parameter", "Drug_Name"],
    var_name = "Time",
    value_name = "Reading"
)

data_melt
```

### <code style="background:yellow;color:black">pivot() method:</code>

* **In pandas, the <code style="background:yellow;color:black">pivot() method</code> is used to reshape a DataFrame, typically by rotating or pivoting it. This method is essentially the inverse operation of the melt function.** 

* **While melt is used to transform a DataFrame from a wide format to a long format, <code 
                                                                                           style="background:yellow;color:black">pivot is used to transform it back from a long format to a wide format.</code>**
* **The basic syntax of the pivot method is: df.pivot(index = None, columns = None, values = None)**
    * index: Column to use as the new DataFrame's index.
    * columns: Column to use as the new DataFrame's columns.
    * values: Column(s) to fill the DataFrame. If not specified, all remaining columns will be used.
    
* **In general, pivot does rows to columns and unpivot does columns to rows.**

* In pandas there is a difference, because, it has given 2 different names to this. Its completely upto the python and pandas developers to name melt and pivot. Concept is same across languages and tools but nomenclature can be different.

**<code style="background:yellow;color:black">Melt -> Convert a bunch of columns to rows**</code> <br>
**<code style="background:yellow;color:black">Pivot -> Does the opposite!</code>**

***

* Lets suppose, for some reason we want to go back to the original dataframe, exactly the way it was before. 

* **Below is the code where we have used <code style="background:yellow;color:black">pivot() method</code> to convert the data_melt to the original dataframe that we had.**

* By saying: <code style="background:yellow;color:black">index = ["Date", "Drug_Name", "Parameter"],</code> we say what columns to retain as it is, they dont need to be changed. 

* From where do we get our columns? To say that we will get our time columns from time column, we will specify like this: <code style="background:yellow;color:black">columns = "Time".</code> 

* From where do we get our values for those columns? To say that we will get our corresponding values from our "reading" column, we will specify like this: <code style="background:yellow;color:black">values = "Reading".</code> 

* **So we have successfully converted our dataset back to its original structure.**

**Now if we observe, this dataframe has become a MULTI-INDEXED DATAFRAME.** Date is the first level of index, within that we have drug name which is the second level of index, and within drug name we have temperature and pressure as third level of index.

**So pivot() by default converted it into a MULTI-INDEXED DATAFRAME.** Difference between original dataframe and pivoted dataframe is there in notes. REFER NOTES FOR MORE CLARITY.

```python
data_melt.pivot(
    index = ["Date", "Drug_Name", "Parameter"],
    columns = "Time",
    values = "Reading"
)
```

**Now if we want the exact original dataframe, we can do the exact same code and add <code style="background:yellow;color:black">.reset_index()</code>. It will make our df back to original df.**

```python
data_melt.pivot(
    index = ["Date", "Drug_Name", "Parameter"],
    columns = "Time",
    values = "Reading"
).reset_index()
```

# <code style="background:yellow;color:black">pivot_table() function:</code>

* **The <code style="background:yellow;color:black">pivot_table() function</code> in pandas is a powerful tool for creating spreadsheet-style pivot tables from a DataFrame.**

* **It allows us to summarize and aggregate data based on one or more key columns and display the results in a tabular form. This function is particularly useful for aggregating and analyzing data when we want to examine relationships between multiple variables.**

* **The <code style="background:yellow;color:black">basic syntax</code> of the pivot_table() function is: pd.pivot_table(data, values = None, index = None, columns = None, aggfunc = 'mean', fill_value = None, margins = False, margins_name = 'All', dropna = True, observed = False)**
    * data: The DataFrame to be used for creating the pivot table.
    * values: The column to aggregate.
    * index: The key column(s) to group by on the rows.
    * columns: The key column(s) to group by on the columns.
    * aggfunc: The aggregation function to apply. Default is 'mean'.
    * fill_value: A scalar value to replace missing values.
    * margins: If True, adds all row and column margins.
    * margins_name: Name to be used for the row/column that contains the totals when margins is True.
    * dropna: If True, do not include columns whose entries are all NaN.
    * observed: If True, only show observed values for categorical groupers.

***

* Now in our example dataframe "data_tidy", we have reached to where we started from. We have made the dataset more presentable, usable and readable. 

* **<code style="background:yellow;color:black">Currently we only have structured the data beautifully. But we want some insights from it.</code>** 

* **Lets suppose we want mean temperature every single day. In other words, we want date wise mean temperature of every drug.** 

* There might be 10 different ways to do this partiular analysis. Here we will talk about one way.

* **When we are trying to aggregate on the basis of date, we need some aggregator function. We want some function to aggregate when we put together multiple values. We do this in groupby also. In this case, we want the aggregator to be average.** 

* **So, <code style="background:yellow;color:black">with the help of pivot_table() function we can do this,</code> which is implemented in the code cell below.**

* **In pivot_table() function, first of all we pass the dataframe i.e data_tidy, index name we want to be "Drug_Name", columns we will take from "Date", Values we will take from "Temperature", and then we will pass the aggregator function, which in this case will be np.mean.** 

* pivot_table() function, very similar to pivot tables in excel or tableau, is a way of summarizing data.

* **This function internally also uses groupby.**

```python
data_tidy.pivot(
    index = ["Drug_Name"],
    columns = "Date",
    values = "Temperature"
)
```

```python
pd.pivot_table(
    data_tidy,
    index = "Drug_Name",
    columns = "Date",
    values = ["Temperature"],
    aggfunc = np.mean
)
```

* **If we want multiple columns, and we want to use minimum aggregator function, we can execute the code below.**

* **This is a direct way of summarizing data without having to write a lengthy groupby code. This way is much shorter in format than groupby.**

* No complex and multiple brackets and chain of functions. Single function which is doing everything for us, which is grouping also, which is aggregating also, which is selecting the values also.

```python
pd.pivot_table(
    data_tidy,
    index = "Drug_Name", # Rows
    columns = "Date", # columns
    values = ["Temperature", "Pressure"], # columns to be aggregated for values
    aggfunc = np.min # aggregator function
)
```
