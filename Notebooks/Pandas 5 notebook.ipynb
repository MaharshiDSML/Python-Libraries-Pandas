# <code style="background:yellow;color:black">NaN and None values:</code>

**In pandas, <code style="background:yellow;color:black">NaN and None are used to represent missing or undefined values in a DataFrame or Series.</code> Ultimately both NaN and None both represent the same thing in terms of data and that is the value is missing or undefined. While they are both used to indicate the absence of data, there are some differences between them:**

**<code style="background:yellow;color:black">NaN (Not a Number):</code>**

* **NaN is a special floating-point value</code> defined by the IEEE floating-point standard. It is commonly used in pandas to represent missing or undefined numerical data.**
* When we perform mathematical operations involving NaN, the result is usually another NaN.
* In pandas, NaN is of the **float type.**

**<code style="background:yellow;color:black">None:</code>**

* **None is a Python object that represents the absence of a value or a null value.</code> It is a general-purpose object and can be used to represent missing values in both numerical and non-numerical (e.g., object, string) data.**
* When working with None, its behavior can vary depending on the operations.
* In pandas, None is often automatically replaced by NaN when working with numeric data.

**<code style="background:yellow;color:black">NaN -> Not a Number -> Value is missing or undefined</code>**<br>
**<code style="background:yellow;color:black">None -> Nothing -> Value is missing or undefined</code>**


* **So what pandas does that is if any column is numerical, then it will automatically keep the missing values as "NaN". And if the column is not numerical, string for example, then pandas will keep the missing values as "None".** 

* Pandas can use them interchangeably because it sees no difference between Nan and None. It is only for presentation purposes, internally they are aliases to one another. So we dont have to remember Nan or None, pandas has created functions which recognises both of them, all that we need to remember is that either NaN or None represent missing values.

** To solve this issue we will be using data_tidy, apply() and groupby() in some form. **

```python
data_tidy
```

* First we groupby the Drug_Name in the data_tidy, then on top of that we calculate the average for temperature.
* Secondly we fill in that average into the missing values. 

* Step 1- Groupby the drug name, calculate the average of temperature.
* Step2- Put these averages into the seperate column called temperature average, and then we will fill missing values using the temperature average column

* So that  temperature average column is going to contain the respective average for that paticular drug across all the times. For this we are going to use apply() method.

* In the function temp_mean(x), "x" is a group and in that entire group, we are adding a new column called "Temp_Avg" of which the column value is equal to the temperature column and the average of it. This is our apply function.

```python
def temp_mean(x):
    # x will be a group
    # In that group, we are setting the value of the entire Temp_Avg column to the average value of the Temperature column
    
    x["Temp_Avg"] = x["Temperature"].mean()
    
    return x
```

We will groupby data_tidy on Drug_Name and then apply apply() method and in that we will call the temp_mean function defined above. "group_keys = False" is just to avoid the jupyter warning. 

```python
data_tidy = data_tidy.groupby(["Drug_Name"], group_keys = False).apply(temp_mean)

data_tidy
```

```python
data_tidy = data_tidy.sort_values(by = "Temp_Avg", ascending = False)

data_tidy
```

**So if we now look at data_tidy, what we have various columns including Temp_Avg. So if we look at this particular <mark>"docetaxel injection"</mark> group, the temperature average for it is <mark>30.387097</mark> and so on.** 	

```python
data_tidy = data_tidy.reset_index()

data_tidy
```

**We have calculated the average temperature but we have not filled the missing values. We will fill them by picking the corresponding average value. Below is the code for that-** 

```python
data_tidy["Temperature"].fillna(data_tidy["Temp_Avg"], inplace = True)
```

**Output of this code below is 0, which means all our NaN or None values of Temperature column have been filled with corresponding Temp_Avg column values.**

```python
data_tidy["Temperature"].isna().sum()
```

**There are 3 steps that we followed and will again follow for pressure-**

* <mark>Step 1</mark>- Write down an apply function.
* <mark>Step 2</mark>- Groupby the data and apply the apply() function to the fresh column 
* <mark>Step 3</mark>- Fill in the missing NaN or None values with corresponding average column values.

```python
# STEP 1

def pressure_mean(x):
    # x will be a group
    # In that group, we are setting the value of the entire Pressure_Avg column to the average value of the Pressure column
    
    x["Pressure_Avg"] = x["Pressure"].mean()
    
    return x
```

```python
# STEP 2

data_tidy = data_tidy.groupby(["Drug_Name"], group_keys = False).apply(pressure_mean)
```

```python
# STEP 3

data_tidy["Pressure"].fillna(data_tidy["Pressure_Avg"], inplace = True)
```

**This code below outputs 0, which means all our NaN or None values of Pressure column have been filled with corresponding Pressure_Avg column values.**

```python
data_tidy["Pressure"].isna().sum()
```

**This code below outputs 0, which means all our NaN or None values of Temperature and Pressure column have been filled with their  corresponding Temp_Avg and Pressure_Avg column values.**

**<mark>Our entire process of filling in the missing values is now COMPLETE.</mark>**

```python
data_tidy.isna().sum()
```
